# Definite Integrals

```python
# Use the definition of the definite integral to evaluate:
# ∫(a, b = 0, 2)(x**2)(dx)
# Use a right-endpoint approximation to generate the Riemann sum.

# Define interval [a, b] and the number of subintervals n

a, b = 0, 2
n = 10000

# Define width of each subinterval
delta_x = (b-a)/n

# Define the right endpoints of each subinterval
x_values = [a + i*delta_x for i in range(1, n+1)]

# Define function to integrate
f = x**2

# Calculate Riemann sum using the right endpoint approximation
riemann_sum = sum([f.subs(x, x_values[i]) * delta_x for i in range(n)])

# Evaluate the definite integral using the limit of the Riemann sum as n approaches ∞
definite_integral = smp.limit(riemann_sum, n, smp.oo)

definite_integral
```

# Average of a Function

```python
# Find the average value of f(x) = x + 1 over [0, 5]

f = x + 1

# Interval limits
a, b = 0, 5

# Calculate definite integral of f(x) over [a, b]
integral = smp.integrate(f, (x, a, b))

# Calculate average value of f(x) over [a, b]
avg_val = (1/(b-a)) * integral

avg_val
```

# Fundamental Theorem of the Calculus: Part One

```python
# Find g(x) = Integral(a, b = 1, x)(1/(t**3+1))(dt)

f_t = 1/(t**3 + 1) # Set equal to function
b_x = x # Set equal to b

b_prime_x = smp.diff(b_x, x)

f_prime_x = f_t.subs(t, b_x) * b_prime_x

f_prime_x
```

```python
# Let F(x) = Integral(a, b = 1, sqrt(x))(sin(t))(dt). Find F'(x)

F_x = smp.sin(t)
b_x = sqrt(x)

b_prime_x = smp.diff(b_x, x)
F_prime_x = F_x.subs(t, b_x) * b_prime_x

F_prime_x
```

```python
# Using the Fundamental Theorem of Calculus with Two Variable Limits of Integration

# Let F(x) = Integral(x, 2x)(t**3)(dt). Find F'(x)

f_t = t**3 # Function

F_x = smp.integrate(f_t, (t, x, 2*x)) # Computes definite integral of f_t with respect to t. Limits set to x (lower limit) and 2*x (upper limit)
F_prime_x = smp.diff(F_x, x) # Find the derivative of F_x

F_prime_x
```

# Integration Formulas / Net Change Theorem

```python
# Integral(a, b = 1, 4)(sqrt(t))(1 + t)(dt). ***Using the Power Rule***

f = smp.sqrt(t)*(1 + t)

# Expand the expression
expanded_f = smp.expand(f)

# Integrate the expanded expression
integral = smp.integrate(expanded_f, t)

# Evaluate the definite integral from 1 to 4
result = integral.subs(t, 4) - integral.subs(t, 1)

result
```

```python
# Given a velocity = v(t)=3t−5 in meters/sec for a particle in motion from time t=0 to time  t=3, find the net displacement of the particle.

v = 3*t - 5

# Integrate the velocity function
s = smp.integrate(v, t)

# Evaluate the definite integral from 0 to 3
net_displacement = s.subs(t, 3) - s.subs(t, 0)

net_displacement
```

```python
# Given v(t) = 3*t - 5 over [0, 3] and net displacement is -3/2, find total distance. 

v = 3*t - 5
s = smp.integrate(v, t)
net_disp = s.subs(t, 3) - s.subs(t, 0) # (-3/2)

# Find time when vel = 0
t_0 = smp.solve(v, t)[0] # (5/3)

# Integrate abs val of vel function over subintervals
tot_dist = smp.integrate(-v, (t, 0, t_0)) + smp.integrate(v, (t, t_0, 3))
tot_dist
```

```python
# Find net displacement and total distance:
# f(t) = (1/3)e^t - 3 over [0, 3]

v = (1/3)*smp.exp(t) - 3
s = smp.integrate(v, t)
net_disp = s.subs(t, 3) - s.subs(t, 0)
# Net Displacement ≈ -2.64

t_0 = smp.solve(v, t)[0]
tot_dist = smp.integrate(-v, (t, 0, t_0)) + smp.integrate(v, (t, t_0, 3))
# Total Distance ≈ 5.21
```

```python
# If engine is started at t = 0 and consumes 5 - t**3 gasoline fir the first hour, how much gas is used in the first hour?
# Integral(a, b = 0, 1)(5 - t**3)(dt)

integral = smp.integrate(5 - t**3)

(integral.subs(t, 0) - integral.subs(t, 1)).evalf()

# Answer is outputed as -4.75. 
# That means that 4.75 gallons of gas are used in 1 hour
```
