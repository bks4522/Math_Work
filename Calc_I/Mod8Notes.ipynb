```python
# Using substitution to find the antiderivative

f_x = 6*x*((3*x**2 + 4)**4)
u_sub = 3*x**2 + 4

# Substitute u into f(x)
f_u = f_x.subs(u_sub, u) / smp.diff(u_sub, x)
# f_u = u**4

# Find the integral w/ respect to u
F_u = smp.integrate(f_u, u)
# F_u = (u**5)/5

# Subsitute x back into integral
F_x = F_u.subs(u, u_sub)

F_x # + C
```

Find the price-demand equation for a brand of toothpase at a supermarket. 

Demand = 50/week at 2.35/tube

Marginal price-demand function = p'(x)

p'(x) = -.015e^-.01x

If 100 tubes sold/week, what should price be?

```python
# Use substitution to find the antiderivative
p_x = (-.015)*smp.exp(-.01*x)
u_sub = (-.01*x)

# Substitute u into p(x)
p_u = p_x.subs(u_sub, u) / smp.diff(u_sub, x)
# Find the integral w/ respect to u
P_u = smp.integrate(p_u, u)
# Subsitute x back into integral
P_x = P_u.subs(u, u_sub)
P_x # + C

# Take the unit price and subtract it by P(50)
# P_x.subs(x, 50) = approx. .91
# 2.35 - .91 = C
C = 2.35 - (P_x.subs(x, 50))

# Now, solve for P(100) + C
P_x.subs(x, 100) + C
```

Rate of growth of bacteria in Petri dish =

q(t) = 3^t

t = time in hours

q(t) = thousands of bacteria per hour

If culture starts w/ 10,000 bacteria, find Q(t) that gives number of bacteria in Petri dish at any time t. 

How many after 2 hours?

```python
# Q(t) = 3**t = (3**t)/ln(3) + C
Q_t = (3**t)/ln(3)

# At t = 0, take 10,000 and subtract Q(0) --- This gives us C
C = 10 - Q_t.subs(t, 0).evalf()

# At t = 2
(Q_t.subs(t, 2) + C).evalf()
```

Suppose growth rate = 

q(t) = 5**t

t = time in hours

q(t) = thousands of bacteria/hour

If culture starts w/ 30,000, find Q(t) for any time t and at t = 3. 

```python
# Q(t) = 5**t = (5**t)/ln(5) + C
Q_t = (5**t)/ln(5)

# At t = 0, take 30,000 and subtract Q(0) --- This gives us C
C = 30 - Q_t.subs(t, 0).evalf()

# At t = 3
(Q_t.subs(t, 3) + C).evalf()
```

Population of flies increases at a rate of:

g(t) = 2e^.02t in flies per day

Initial population of flies is 100, how many after 10 days?

```python
# Using substitution to find a definite integral
g_t = 2*smp.exp(.02*t)
u_sub = .02*t
# Substitute the variables
g_u = g_t.subs(u_sub, u) / smp.diff(u_sub, t)
g_u = g_u.subs(t, smp.sqrt(u))
# Limits
lower_limit = u_sub.subs(t, 0)
upper_limit = u_sub.subs(t, 10)
# Evaluate the integral
integral = smp.integrate(g_u, (u, lower_limit, upper_limit))

# Initial population + integral
100 + integral
```
