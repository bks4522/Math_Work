# Box Maximization

BOX

An open-top box is to be made from a  24
  in. by  36
  in. piece of cardboard by removing a square from each corner of the box and folding up the flaps on each side. What size square should be cut out of each corner to get a box with the maximum volume?

  Step 1: Let  x
  be the side length of the square to be removed from each corner. Then, the remaining four flaps can be folded up to form an open-top box. Let  V
  be the volume of the resulting box.
  Thus, 
  p1 = 36-2x,
  p2 = 24-2x, and
  p3 = x

  Box Volume = V = L*W*H. p3 = H, L = 36-2x, and W = 24-2x

  Thus, V(x) = (p1)(p2)(p3)

  For a box, the side length of the square cannot be greater than or equal to half the length of the shorter side (24). Thus, find over the interval (0, 12). 

  V(x) = 0 at endpoints and V(x) . 0 for 0 < x < 12. 

  V'(x) = 12x^2 - 240x + 864. Solving for 0 to find critical points yields:

  x = 10±2sqrt(7)

```python
p1 = 36-2*x # L
p2 = 24-2*x # W
p3 = x # H

volume = smp.expand(p1 * p2 * p3) # V = LWH
volume
smp.diff(volume, x)

# Finding critical points (cp):
cp = smp.solve(smp.diff(volume, x), x) # 10 ± 2sqrt(7)
cp
```

```python
# Plugging in critical points into volume
volume.subs(x, cp[0]).expand()
```

# Minimizing Travel Time

An island is  2mi
  due north of its closest point along a straight shoreline. A visitor is staying at a cabin on the shore that is  6mi
  west of that point. The visitor is planning to go from the cabin to the island. Suppose the visitor runs at a rate of  8mph
  and swims at a rate of  3mph.
  How far should the visitor run before swimming to minimize the time it takes to reach the island?

  Step 1: Let  x
  be the distance running and let  y
  be the distance swimming. Let  T
  be the time it takes to get from the cabin to the island.

  Step 2: The problem is to minimize  T.
 
Step 3: To find the time spent traveling from the cabin to the island, add the time spent running and the time spent swimming. Since Distance  =
  Rate  ×
  Time  (D=R×T),
  the time spent running is

Trunning = Drunning/Rrunning=x/8,
 
and the time spent swimming is

Tswimming=Dswimming/Rswimming=y/3.
 
Therefore, the total time spent traveling is

T=x/8+y/3.

Step 4: From the figure, the line segment of  y
  miles forms the hypotenuse of a right triangle with legs of length  2mi
  and  6−xmi.
  Therefore, by the Pythagorean theorem,  2^2+(6−x)^2=y^2,
  and we obtain  y = sqrt((6-x)^2 + 4)
  Thus, the total time spent traveling is given by the function

  T(x) = x/8 + sqrt((6-x)^2 + 4)/3

  Step 5: From the figure, we see that  0≤x≤6.
  Therefore,  [0,6]
  is the domain of consideration.

  Step 6: Since T(x)
 is a continuous function over a closed, bounded interval, it has a maximum and a minimum. Let’s begin by looking for any critical points of T
 over the interval  [0,6].

 T'(x) = (6-x)/3sqrt((6-x)^2 + 4)

 Setting T'(x) = 0 yields x = 1/8. 

 This, if 1/8 = (6-x)/3sqrt((6-x)^2 + 4), then it follows that 3sqrt((6-x)^2 + 4 = 8(6-x)


This yields x = 6 - 6/sqrt(55). In Colabs, it outputs slightly differently. 


 Since  x = 6 - 6/sqrt(55)
  does satisfy that equation, we conclude that  x = 6 - 6/sqrt(55)
  is a critical point, and it is the only one. To justify that the time is minimized for this value of  x,
  we just need to check the values of  T(x)
  at the endpoints  x=0
  and  x=6,
  and compare them with the value of  T(x)
  at the critical point  x = 6−6/sqrt(55). 
  We find that  T(0)≈2.108h
  and  T(6)≈1.417h,
  whereas  T(6−6/sqrt(55))≈1.368h.
  Therefore, we conclude that  T
  has a local minimum at  x≈5.19
  mi.

```python
# Expression
expr = 3*sqrt((6-x)**2 + 4) - (8*(6-x))

# Solving for x to find cp
cp = smp.solve(expr, x) # yields: 6 - 6*sqrt(55)/55. The "6+" is not in the domain. Thus, this is a critical point.  
cp

# Numerical Value
numcp = smp.N(cp[0]) # x ≈ 5.19 --- Local Minimum
numcp
```

# Maximizing Revenue

Owners of a car rental company have determined that if they charge customers  p
  dollars per day to rent a car, where  50 ≤ p ≤ 200,
  the number of cars  n
  they rent per day can be modeled by the linear function  n(p) = 1000 − 5p.
  If they charge  $50
  per day or less, they will rent all their cars. If they charge  $200
  per day or more, they will not rent any cars. Assuming the owners plan to charge customers between $50 per day and  $200
  per day to rent a car, how much should they charge to maximize their revenue?

Step 1: Let  p
  be the price charged per car per day and let  n
  be the number of cars rented per day. Let  R
  be the revenue per day.

Step 2: The problem is to maximize  R.
 
Step 3: The revenue (per day) is equal to the number of cars rented per day times the price charged per car per day—that is,  R = n × p.

Step 4: Since the number of cars rented per day is modeled by the linear function  n(p)=1000−5p,
  the revenue  R
  can be represented by the function



R(p) = n × p = (1000−5p)p = −5p2+1000p


 
Step 5: Since the owners plan to charge between  50 dollars per car per day and  $200
  per car per day, the problem is to find the maximum revenue  R(p)
  for  p
  in the closed interval  [50,200].

  Step 6: Since  R
  is a continuous function over the closed, bounded interval  [50,200],
  it has an absolute maximum (and an absolute minimum) in that interval. To find the maximum value, look for critical points. The derivative is  R′(p)=−10p+1000.
  Therefore, the critical point is  p=100
  When  p=100,
   R(100)=$50,000.
  When  p=50,
   R(p)=$37,500.
  When  p=200,
   R(p)=$0.

  Therefore, the absolute maximum occurs at  p=$100.
  
  The car rental company should charge  $100
  per day per car to maximize revenue as shown in the following figure.

_______________________________________________________________

Linear Function = 1000 - 5p

R is a continuous function over the closed, bounded interval [50, 200]. It has an absolute max/min on the interval. 

To find max, look for critical points (cp). 

R'(p) = -10p + 1000

cp = 100

R(100) = 50,000

R(50) = 37,500

R(200) = 0

Thus, absolute maximum is p = 100. 

```python
# Revenue = number of cars rented per day * price charged per car per day
# R = n * p

# R(p) = n * p = (1000 - 5p)p
# Thus:
# R(p) = -5p**2 + 1000p

# $50 /car/day and $200 /car/day
# Thus, find revenue over [50, 200]

Rp = -5*p**2 + 1000*p
Rpprime = smp.diff(Rp, p)

# Solve for critical points

cp = smp.solve(Rpprime, p) # 100

# Solve for R(100)
Rp.subs(p, cp[0])

# Solve for R(50)
Rp.subs(p, 50)

# Solve for R(200)
Rp.subs(p, 200)
```

# Maximizing the Area of an Inscribed Rectangle

Rectangle is to be inscribed in the ellipse:

(x^2/4) + y^2 = 1

What should the dimensions of the rectangle be to maximize its area? What is the maximum area?

Step 1: For a rectangle to be inscribed in the ellipse, the sides of the rectangle must be parallel to the axes. Let  L
  be the length of the rectangle and  W
  be its width. Let  A
  be the area of the rectangle.

  Step 2: The problem is to maximize  A.
 
Step 3: The area of the rectangle is  A=LW.

Step 4: Let (x, y) be the corner of the rectangle that lies int he 1st quadrant. We can write L = 2x and W = 2y

Since (x^2/4) + y^2 = 1 and y > 0, we have: 

y = sqrt(1-(x^2/4))

Thus:

A = LW = (2x)(2y) = 4x*sqrt(1-(x^2/4)) = 

2x*sqrt(4 - x^2)

Step 5: From Figure 4.68 , we see that to inscribe a rectangle in the ellipse, the  x
  -coordinate of the corner in the first quadrant must satisfy  0 < x < 2.
  Therefore, the problem reduces to looking for the maximum value of  A(x)
  over the open interval  (0,2).
  Since  A(x)
  will have an absolute maximum (and absolute minimum) over the closed interval  [0,2],
  we consider  A(x) = 2x*sqrt(4 - x^2)
  over the interval  [0,2].
  If the absolute maximum occurs at an interior point, then we have found an absolute maximum in the open interval

  _______________________________

  At endpoints x = 0 and x = 2, A(x) = 0. For 0 < x < 2, A(x) > 0. 

  Thus, the max occurs on the critical point. 

  Taking the derivative:

  A'(x) = (8-4x^2)/(sqrt(4 - x^2)). 

  Solving for A'(x) x = ±sqrt(2)

  -sqrt(2) IS NOT on the interval. sqrt(2) = Critical Point

  Thus, dimensions:

  L = 2x = 2sqrt(2)

  W = 2y = 2/sqrt(2) = sqrt(2)

  A = LW = 4

```python
# L = 2x
# W = 2y

# (x^2/4) + y^2 = 1 and y > 0:
y = sqrt(1 - (x**2/4))

# Since A = LW:

Aofx = (2*x)*(2*y)

# A'(x) = 
Aofxprime = smp.diff(Aofx, x)

# Find critical point
cp = smp.solve(Aofxprime, x)

# Plugging cp into y
solve = y.subs(x, cp[1]) # sqrt(2)/2 is the same as 1/sqrt(2)

L = 2*solve
W = 2/solve

# Solve for area
L*W
```

# Minimizing Surface Area (Interval not closed / is unbounded)

A rectangular box with a square base, an open top, and a volume of  216
  in^3 is to be constructed. What should the dimensions of the box be to minimize the surface area of the box? What is the minimum surface area?

  Step 1: Draw a rectangular box and introduce the variable  x
  to represent the length of each side of the square base; let  y
  represent the height of the box. Let  S
  denote the surface area of the open-top box.

  Step 2: We need to minimize the surface area. Therefore, we need to minimize  S.

  Step 3: Since the box has an open top, we need only determine the area of the four vertical sides and the base. The area of each of the four vertical sides is  x*y.
  The area of the base is  x^2.
  Therefore, the surface area of the box is

S = 4xy + x^2.

Step 4: Since the volume of this box is x^2(y)
  and the volume is given as  216 in^3,
  the constraint equation is

x^2(y) = 216

Solving the constraint equation for y yeilds:

y = 216/x^2

Thus:

S(x) = 4x(216/x^2) + x^2

Thus: S(x) = 864 + x^2

Step 5: Since we are requiring that  x^2(y) = 216,
  we cannot have  x = 0.
  Therefore, we need  x > 0.
  On the other hand,  x
  is allowed to have any positive value. Note that as  x
  becomes large, the height of the box  y
  becomes correspondingly small so that  x^2(y) = 216.
  Similarly, as  x
  becomes small, the height of the box becomes correspondingly large. We conclude that the domain is the open, unbounded interval  (0,∞).
  Note that, unlike the previous examples, we cannot reduce our problem to looking for an absolute maximum or absolute minimum over a closed, bounded interval. However, in the next step, we discover why this function must have an absolute minimum over the interval  (0,∞).

  Step 6: As x→0+, S(x)→∞. As x→∞, S(x)→∞. 

  S is a continuous function that pproaches infinity at the ends, it must have an absolute min at some x is an element in (0, ∞). 

  S'(x) = (-864/x^2) + 2x

  Thus, when S'(x) = 0, 2x = 864/x^2

  Solving for x yields:

  x = cbrt(432) or 6cbrt(2) [Critical Point]

  Thus, the critical point 6cbrt(2) = absolute minimum. 

  Plugging in critical point into x^2(y) = 216:

  y = 216/(x)^2 = 3cbrt(2)

  Thus: 

  x = 6cbrt(2)

  y = 3cbrt(2)

  Solving for S(x) yields: 108cbrt(4) in^2

```python
# Rectangular box w/ square base and open top
# Volume of 216 in^3
# Since it has an open top, we need to find the four sides and the base. 
# Base area = x^2

# S = 4*x*y + x**2

# Thus, (x^2)y = 216

y = 216 / x**2

Sofx = 4*x*(216/x**2) + x**2
Sofx # x^2 + 864/x

# S'(x)
Sofxprime = smp.diff(Sofx, x)

cp = smp.solve(Sofxprime, x) # Critical Point/absolute minimum: 6*cbrt(2)

216 / (cp[0])**2 # y value

# Thus, x = 6*cbrt(2) and y = 3*cbrt(2)

Sofx.subs(x, cp[0]) # 108*cbrt(4) in^2
```

# L’Hôpital’s Rule

When you can use LH Rule:

1. Limit of quotient is the form 0/0 or ±∞/±∞. To find this, plug in the limit into x. 

2. Both numerator and denominator are differentiable functions (i.e., have a derivative). To find this, take the derivative of num/den. 

3. Limit of the quotient of the derivatives exists or is ±∞

This means that LH's Rule is: limit of the quotient of the functions is equal to the limit of the quotient of their derivatives

CANNOT use LH Rule if:

1. Limit of the quotient is not of the form 0/0 or ±∞/±∞

2. One or both functions in the numerator or denominator are not differentiable at the point of interest.

3. Limit of the quotient of the derivatives DNE or keeps oscillating b/t indeterminate forms.

# Newton's Method

```python
# Use Newton's Method to approximate root of f(x) = x**3 - 3*x + 1 over the interval [1, 2]. Let x0 = 2. Find x1, x2, x3, x4, x5

fx = x**3 - 3*x + 1
fpx = smp.diff(x**3 - 3*x + 1, x)

x1 = 2 - (fx.subs(x, 2)/(fpx.subs(x, 2)))
x2 = x1 - (fx.subs(x, x1)/(fpx.subs(x, x1)))
x3 = x2 - (fx.subs(x, x2)/(fpx.subs(x, x2)))
x4 = x3 - (fx.subs(x, x3)/(fpx.subs(x, x3)))
x5 = x4 - (fx.subs(x, x4)/(fpx.subs(x, x4)))
x6 = x5 - (fx.subs(x, x5)/(fpx.subs(x, x5)))
```

```python
# Use Newton's Method to approx root of f(x) = x**3 - 8*x + 1 over [0, 1]. Let x0 = 0. Find x1 and x2

fx = x**3 - 8*x + 1
fpx = smp.diff(x**3 - 8*x + 1, x)

x1 = 0 - (fx.subs(x, 0)/(fpx.subs(x, 0)))
x2 = x1 - (fx.subs(x, x1)/(fpx.subs(x, x1)))
```

```python
# Initial Value Problem

# dy/dx = sinx, y(0) = 5

smp.integrate(smp.sin(x))
# -cos(x) + C

# Look for solution y that satisfies y(0) = 5. This means we need C such that: 

# -cos(0) + C = 5
5 + smp.cos(0)
# C = 5 + cos(0) = 6

# Therefore, solution to initial value problem is: y = -cos(x) + 6
```

```python
# Initial Value Problem

# dy/dx = 3x**-2, y(1) = 6

smp.integrate(3*x**-2)
# -3/x + C

# Look for solution y that satisfies y(1) = 6. This means we need C such that: 

#-3/1 + C = 6
6 + (3/1)
# C = 6 + (3/1) = 9

# Thus, solution to initial value problem is: y = (-3/x) + 2
```

```python
# Inital Value Problem

# Find an antiderivative of F(x) with F'(x) = f(x) = 9 + 24*x**3 + 18*x**5 and F(1) = 0

f = 9 + 24*x**3 + 18*x**5

# Find indefinite integral of f(x)
F = smp.integrate(f, x)

# Add constant + C 
C = smp.symbols("C")
F = F + C

# Find C such that F(1) = 0
C_value = smp.solve(F.subs(x, 1), C)

# Sub C back into F(x)

F = F.subs(C, C_value[0])
F
```

A car is traveling at the rate of  88
  ft/sec  when the brakes are applied. The car begins decelerating at a constant rate of  15
  ft/sec ^2 .

How many seconds elapse before the car stops?

How far does the car travel during that time?

```python
# t = time in seconds
# a(t) = acceleration of the car in ft/s**2
# v(t) = velocity of the car in ft/s**2
# s(t) = car's position in ft beyond the point where the brakes are applied at time t. 

# Traveling at 88 ft/s
# Therefore, initial velocity v(0) = 88 ft/s
# Since car is decelerating: a(t) = -15 ft/s**2

# v'(t) = -15

# v'(t) = -15 and v(0) = 88

# v(t) = -15t + C

# Since v(0) - 88, C = 88, thus the velocity fuction is: v(t) = -15t + 88

# Solving for -15t + 88 = 0
smp.solve(-15*t + 88, t)
# t = 88/15

# v(t) is the derivative of the position s(t). Initial position = s(0) = 0. 
# Therefore, we need to solve s'(t) = -15t + 88, s(0) = 0. 
smp.integrate(-15*t + 88) # + C

# Since s(0) = 0, C = 0.
# Position function: 
st = (-15/2)*t**2 + 88*t
# Plugging in s(88/15)
st.subs(t, (88/15))
```

  ft/sec**2.

 

How long does it take for the car to stop?

```python
# v(0) = 50 ft/s
# Decelerating: a(t) = -17 ft/s**2

# v'(t) = -17 and v(0) = 50

# v(t) = -17t + C

# v(t) = -17t + 50
smp.solve(-17*t + 50)
# t = 50/17

# Solve s'(t) = -17t + 50, s(0) = 0
smp.integrate(-17*t + 50) # + C

# Since s(0) = 0, C = 0
# Position function:
st = (-17/2)*t**2 + 50*t
# Plugging in s(50/17)
st.subs(t, (50/17))
```

```python
# Approximating the Area under a Curve

# Use left-endpoint and right-endpoint approx to approx the area under the curve:
# f(x) = x**2 on the interval [0, 2]; use n = 4

f = x**2 # Function
a, b = 0, 2 # Interval
n = 4 # Subintervals
delta_x = (b-a)/n # Delta X

left_endpoints = [a + i * delta_x for i in range(n)] # Creates a list of left endpoints for each subinterval. 
left_area = sum(f.subs(x, left_endpoint) * delta_x for left_endpoint in left_endpoints) # Calculates the approximations
#left_area

midpoints = [a + (i + .5) * delta_x for i in range(n)]
midpoint_area = sum(f.subs(x, midpoint) * delta_x for midpoint in midpoints)
#midpoint_area

right_endpoints = [a + (i + 1) * delta_x for i in range(n)] # Creates a list of left endpoints for each subinterval.
right_area = sum(f.subs(x, right_endpoint) * delta_x for right_endpoint in right_endpoints) # Calculates the approximations
#right_area

print("Left area:", left_area, ", Midpoint", midpoint_area, "Right area:", right_area)
```
