Which polar coordinate pair labels: (-3, pi/4)?

```python
# Given polar coordinates
r, theta = -3, smp.pi/4

# Convert to positive radius
r = abs(r)

# Adjust the angle
theta1 = theta - smp.pi
theta2 = theta + smp.pi

# Make sure the angles are between 0 and 2pi
if theta1 < 0:
    theta1 += 2*smp.pi

print((r, theta1))
```

# 3

Find the polar coordinates, 0 <= Theta < 2pi and r >= 0:

A: (4*sqrt(3), 4)

B: (-2, 2*sqrt(3))

C: (-4, -4*sqrt(3))

```python
# A

# Given Cartesian coordinates
x, y = 4*smp.sqrt(3), 4

# Convert to polar coordinates
r = smp.sqrt(x**2 + y**2)
theta = smp.atan2(y, x)

# Make sure the angle is between 0 and 2pi
if theta < 0:
    theta += 2*smp.pi

print((r, theta))
```

```python
# B

# Given Cartesian coordinates
x, y = -2, 2*smp.sqrt(3)

# Convert to polar coordinates
r = smp.sqrt(x**2 + y**2)
theta = smp.atan2(y, x)

# Make sure the angle is between 0 and 2pi
if theta < 0:
    theta += 2*smp.pi

print((r, theta))
```

```python
# C

# Given Cartesian coordinates
x, y = -4, -4*smp.sqrt(3)

# Convert to polar coordinates
r = smp.sqrt(x**2 + y**2)
theta = smp.atan2(y, x)

# Make sure the angle is between 0 and 2pi
if theta < 0:
    theta += 2*smp.pi

print((r, theta))
```
