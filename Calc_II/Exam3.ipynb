```python
# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Calculate the corresponding y values
y = 3*x + 9

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y)

# Set the aspect of the plot to be equal, so the scale is the same on the x and y axes
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = 3x + 9')

# Show the plot
plt.show()
```
