Calculus/Statistics Functions

```python
# np.random.randn generates random numbers with a mean of 0 and a standard deviation of 1. 
# 2* expand the standard deviation by 2
# + 10 shifts the mean of the random numbers to the right by 10

a1 = 2*np.random.randn(10000) + 10
a1
np.mean(a1)
np.std(a1) # std = standard deviation
np.percentile(a1, 20) # How many values are less than passed argument. Ex: If passed arg = 20, then approx 8.3% of numbers will be 20 or less
#np.cumsum(np.array([1, 2, 3, 4]))
```

```python
# Integrals and Derivatives

x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)

dydx = np.gradient(y, x)

y_int = np.cumsum(y)*(x[1]-x[0])

plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

```python
# Systems of Equations

# 3x + 2y + z = 4
# 5x - 5y + 4z = 3
# 6x + z = 0

A = np.array([[3, 2, 1], [5, -5, 4], [6, 0, 1]])
c = np.array([4, 3, 0])

np.linalg.solve(A, c) # Ouputs x, y, z values
```

```python
# Finding Eigenvalues

A = np.array([[4, 2, 2], [2, 4, 2], [2, 2, 4]])

np.linalg.eig(A)

w, v = np.linalg.eig(A)

w
v1 = v[:,0]
v1
```

Q1: Let f(x, y) = (e^-(x^2+y^2)) * sin(x) for -2  x  2 and -2  y  2

1. Make a contour plot of f
2. Find the volume |f(x, y)| in the specified x and y range
3. Find the volume |f(x, y)| only in the region where sqrt(x^2 + y^2) > .5

```python
# 1 -- Meshgrid

x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)
xv, yv = np.meshgrid(x, y)
f = np.exp(-(xv**2 + yv**2)) * np.sin(xv)

plt.contourf(xv, yv, f, levels = 100)
plt.colorbar()
```

```python
# 2

np.abs(f.ravel()).sum() * np.diff(x)[0] * np.diff(y)[0] # Multiplied by dx and dy
```

```python
# 3 

np.abs(f[xv**2 + yv**2 > 0.5**2].ravel()).sum() * np.diff(x)[0] * np.diff(y)[0]
```
