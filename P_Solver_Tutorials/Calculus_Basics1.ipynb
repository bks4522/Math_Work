```python
# Implicit Differentiation

eq1 = x**2 + y**2 - ((1/6)*x**2 + (1/6)*y**2 - y)**2
point = (1, 0)

# Partial derivatives with respect to x and y
dx = smp.diff(eq1, x)
dy = smp.diff(eq1, y)

# Evaluate the partial derivatives at the given point
dx_val = dx.subs([(x, point[0]), (y, point[1])])
dy_val = dy.subs([(x, point[0]), (y, point[1])])

# Calculate the slope of the tangent line
slope = -dx_val/dy_val

# Use the point-slope form of the line to get the equation of the tangent line
y2 = slope*(x - point[0]) + point[1]

# Display the equation of the tangent line
print(f"The equation of the tangent line at ({point[0]}, {point[1]}) is y = {smp.simplify(y2)}")
```

# Antiderivatives

```python
# Anti-deriviative of: smp.csc(x)*smp.cot(x)
# + c to final answer

smp.integrate(smp.csc(x)*smp.cot(x), x)
```

```python
# Anti-derivative of: 4*smp.sec(3*x)*smp.tan(3*x)
# + c to final answer

smp.integrate(4*smp.sec(3*x)*smp.tan(3*x), x)
```

```python
# Anti-derivative of : 2/smp.sqrt(1-x**2) - 1/x**smp.Rational(1,4)
# + c to final answer

smp.integrate(2/smp.sqrt(1-x**2) - 1/x**smp.Rational(1,4), x)
```

```python
# Anti-derivative of: (1+smp.sqrt(x))**smp.Rational(1,3) / smp.sqrt(x)
# + c to final answer

smp.integrate((1+smp.sqrt(x))**smp.Rational(1,3) / smp.sqrt(x), x)
```

```python
# Anti-derivative of: x*(1-x**2)**smp.Rational(1/4)
# + c to final answer

smp.integrate(x*(1-x**2)**smp.Rational(1/4), x)
```

```python
# Anti-derivative of: (2*x - 1)*smp.cos(smp.sqrt(3*(2*x-1)**2 + 6)) / smp.sqrt(3*(2*x-1)**2 + 6)
# + c to final answer

smp.integrate((2*x - 1)*smp.cos(smp.sqrt(3*(2*x-1)**2 + 6)) / smp.sqrt(3*(2*x-1)**2 + 6), x)
```

# Initial Value Problems

```python
# Given dy/dx = 8x + csc**2(x) with y(pi/2) = -7, solve for y(x)

integral = smp.integrate(8*x + smp.csc(x)**2, x)

# integral

# integral.subs(x, smp.pi/2)

# Set C = -pi**2 and - 7

C = - integral.subs(x, smp.pi/2) - 7

# C

y = integral + C

# y

# To check: y.subs(x, smp.pi/2)

y.subs(x, smp.pi/2)
```

# Sequences and Series

```python
n = smp.symbols('n')
```

```python
# smp.Sum(6/4**n, (n, 0, smp.oo))

smp.Sum(6/4**n, (n, 0, smp.oo)).doit()
```

```python
# smp.Sum(2**(x+1) / 5**n, (n, 0, smp.oo))

smp.Sum(2**(n+1) / 5**n, (n, 0, smp.oo)).doit()
```

```python
# smp.Sum(smp.atan(n) / n**smp.Rational(11, 10), (n, 1, smp.oo))

# When .doit() doesn't do it: approximate with .n()

smp.Sum(smp.atan(n) / n**smp.Rational(11, 10), (n, 1, smp.oo)).n()
```

```python
# smp.Sum((1 + smp.cos(n)) / n, (n, 1, smp.oo))

# Approximating with .n() DOES NOT yield the right answer. There is no answer. 

smp.Sum((1 + smp.cos(n)) / n, (n, 1, smp.oo)).n()
```

```python
# However, if n is replaced by n**2 in the denominator, we can find an answer:

smp.Sum((1 + smp.cos(n)) / n**2, (n, 1, smp.oo)).n()
```
