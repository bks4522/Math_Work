Let x and y be functions of t. 

Find the general solution of the system of equations below by first converting the system into​ second-order differential equations involving only y and only x. 

Find a particular solution for the initial conditions. 

Use a computer system or graphing calculator to construct a direction field and typical solution curves for the given system:

x' = 16y

x(0) = 2

```python
# Define the system of equations
def system(v, t):
    x, y = v
    dxdt = 16*y
    dydt = -4*x
    return [dxdt, dydt]

# Initial conditions
v0 = [2, 4]  # x(0) = 2, y(0) = 4

# Time points
t = np.linspace(0, 1, 1000)

# Solve the system
v = odeint(system, v0, t)

# Extract the solutions
x = v[:, 0]
y = v[:, 1]

# Find the particular solution
x_particular = x[-1]
y_particular = y[-1]

# Print the particular solution
print("Particular solution:")
print(f"x(1) = {x_particular}") # 2cos(8t) + 8sin(8t)
print(f"y(1) = {y_particular}") # 4cos(8t) - sin(8t)
```

# 9

Apply the eigenvalue method to find a general solution of the given system.

Find the particular solution corresponding to the given initial values.

x'_1 = 3x_1 + 4x_2

x'_2 = 3x_1 + 2x_2

x_1 (0) = x_2 (0) = 1

What is the general solution in matrix form of x(t)?

```python
# Define the time variable
t = np.linspace(0, 1, 1000)

# Define the functions for x1(t) and x2(t)
x1 = (3 + 4*np.exp(7*t))/(7*np.exp(t)) + 4*(-1 + np.exp(7*t))/(7*np.exp(t))
x2 = 3*(-1 + np.exp(7*t))/(7*np.exp(t)) + (4 + 3*np.exp(7*t))/(7*np.exp(t))

# Plot x1(t) and x2(t)
plt.figure(figsize=(10, 5))
plt.plot(t, x1, label='x1(t)')
plt.plot(t, x2, label='x2(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Solution of the system of differential equations')
plt.grid(True)
plt.show()
```

Apply the eigenvalue method to find a general solution of the system.

x'_1 = 4x_1 + 4x_2 + x_3

x'_2 = -6x_1 - 6x_2 - 4x_3

x'_3 = 6x_1 + 6x_2 + 4x_3

What is the general solution in matrix form?

# MATLAB Code

```python
# Define the system in terms of a function
def f(X, t):
    x1, x2 = X
    dx1dt = -5*x1 + x2
    dx2dt = 4*x1 - 2*x2
    return [dx1dt, dx2dt]

# Create an x1 and x2 grid
x1range = np.linspace(-1, 1, 20)
x2range = np.linspace(-1, 1, 20)
x1, x2 = np.meshgrid(x1range, x2range)

# Compute the direction field on the grid
u, v = np.zeros(x1.shape), np.zeros(x2.shape)
NI, NJ = x1.shape

for i in range(NI):
    for j in range(NJ):
        x = x1[i, j]
        y = x2[i, j]
        yprime = f([x, y], 0)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]

# Normalize the arrows for better visualization
N = np.sqrt(u**2+v**2)  
u /= N
v /= N

# Create the direction field plot
plt.quiver(x1, x2, u, v, color='blue')

# Solve the system for the initial conditions and plot the resulting trajectory
t = np.linspace(0, 10, 500)
X0 = [0.5, 0.3]
X = odeint(f, X0, t)
plt.plot(X[:, 0], X[:, 1], 'r')

# Labels, title and limits
plt.title('Phase Portrait')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.grid()
plt.show()
```
