```python
import sympy as smp
from sympy import *
import numpy as np
from numpy import *
import scipy as sp
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d
from scipy.integrate import odeint
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])

x, y, t, n, a, b, C, c, r, u, s = smp.symbols('x y t n a b C c r u s')
f, g, m, h = smp.symbols('f g m h', cls=smp.Function)
```

```python
# C

# Create a time array
t = np.linspace(0, 10, 1000)

# Compute the population at each time point
y = (800 * np.exp(0.4 * t)) / (3 + np.exp(0.4 * t))

# Plot the population over time
plt.plot(t, y)

# Label the axes
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')

# Turn on the grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# E

# Define the logistic function
def logistic(P, t):
    return 0.4*P*(1 - P/800)

# Initial condition
P0 = 200

# Time array
t = np.linspace(0, 10, 1000)

# Solve the differential equation
P = odeint(logistic, P0, t)

# Plot the solution
plt.plot(t, P)
plt.title('Solution of Logistic Population Model')
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')
plt.grid(True)
plt.show()
```

```python
# D

# Create a time array
t = np.linspace(0, 1, 1000)

# Compute the charge at each time point
y = (24*np.cos(300*t) + np.sin(300*t)) / (12*np.exp(10*t))

# Plot the charge over time
plt.plot(t, y)

# Label the axes
plt.xlabel('Time (t)')
plt.ylabel('Charge (Q)')

# Turn on the grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# E

# Define the RLC circuit function
def RLC(t, Q):
    Q, dQdt = Q
    return [dQdt, (1/5)*(-100*dQdt - 450500*Q)]

# Initial conditions
Q0 = [2, 5]

# Time span
t_span = [0, 1]

# Solve the differential equation
sol = solve_ivp(RLC, t_span, Q0, t_eval=np.linspace(0, 1, 1000))

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.title('Solution of RLC Circuit')
plt.xlabel('Time (t)')
plt.ylabel('Charge (Q)')
plt.grid(True)
plt.show()
```

```python
# E

# Define the spring-mass-damper system function
def spring_mass_damper(t, y):
    y, dydt = y
    return [dydt, (1/5)*(25*np.sin(10*t) - 30*dydt - 49*y)]

# Initial conditions
y0 = [1.5, 0]

# Time span
t_span = [0, 20]

# Solve the differential equation
sol = solve_ivp(spring_mass_damper, t_span, y0, t_eval=np.linspace(0, 20, 1000))

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.title('Solution of Spring-Mass-Damper System')
plt.xlabel('Time (t)')
plt.ylabel('Location (y)')
plt.grid(True)
plt.show()
```
