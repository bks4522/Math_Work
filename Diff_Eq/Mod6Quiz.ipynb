```python
# C

# Create a time array
t = np.linspace(0, 10, 1000)

# Compute the population at each time point
y = (800 * np.exp(0.4 * t)) / (3 + np.exp(0.4 * t))

# Plot the population over time
plt.plot(t, y)

# Label the axes
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')

# Turn on the grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# E

# Define the logistic function
def logistic(P, t):
    return 0.4*P*(1 - P/800)

# Initial condition
P0 = 200

# Time array
t = np.linspace(0, 10, 1000)

# Solve the differential equation
P = odeint(logistic, P0, t)

# Plot the solution
plt.plot(t, P)
plt.title('Solution of Logistic Population Model')
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')
plt.grid(True)
plt.show()
```

```python
# D

# Create a time array
t = np.linspace(0, 1, 1000)

# Compute the charge at each time point
y = (24*np.cos(300*t) + np.sin(300*t)) / (12*np.exp(10*t))

# Plot the charge over time
plt.plot(t, y)

# Label the axes
plt.xlabel('Time (t)')
plt.ylabel('Charge (Q)')

# Turn on the grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# E

# Define the RLC circuit function
def RLC(t, Q):
    Q, dQdt = Q
    return [dQdt, (1/5)*(-100*dQdt - 450500*Q)]

# Initial conditions
Q0 = [2, 5]

# Time span
t_span = [0, 1]

# Solve the differential equation
sol = solve_ivp(RLC, t_span, Q0, t_eval=np.linspace(0, 1, 1000))

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.title('Solution of RLC Circuit')
plt.xlabel('Time (t)')
plt.ylabel('Charge (Q)')
plt.grid(True)
plt.show()
```

```python
# E

# Define the spring-mass-damper system function
def spring_mass_damper(t, y):
    y, dydt = y
    return [dydt, (1/5)*(25*np.sin(10*t) - 30*dydt - 49*y)]

# Initial conditions
y0 = [1.5, 0]

# Time span
t_span = [0, 20]

# Solve the differential equation
sol = solve_ivp(spring_mass_damper, t_span, y0, t_eval=np.linspace(0, 20, 1000))

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.title('Solution of Spring-Mass-Damper System')
plt.xlabel('Time (t)')
plt.ylabel('Location (y)')
plt.grid(True)
plt.show()
```

# Mod 6-1

```python
# Define the system of first-order differential equations
def system(t, y):
    dydt = np.zeros(2)
    dydt[0] = y[0] + 2*y[1]
    dydt[1] = 3*y[0] + 2*y[1]
    return dydt

# Define the time points
t = np.linspace(0, 5, 500)

# Define the initial conditions
y0 = [1, 1]

# Solve the system of differential equations
sol = solve_ivp(system, [0, 5], y0, t_eval=t)

# Plot the solution
plt.figure()
plt.plot(sol.t, sol.y[0], '-', label='x1(t)')
plt.plot(sol.t, sol.y[1], ':', label='x2(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()
```

# Mod 6 - 3

```python
# Define the system of first-order differential equations
def system(t, y):
    dydt = np.zeros(2)
    u = 1 if t >= 3 else 0
    dydt[0] = y[0] - 15*y[1] + 20*np.cos(6*np.pi*t)*u
    dydt[1] = 2*y[0] - 5*y[1]
    return dydt

# Define the time points
t = np.linspace(0, 10, 1000)

# Define the initial conditions
y0 = [10, 0]

# Solve the system of differential equations
sol = solve_ivp(system, [0, 10], y0, t_eval=t)

# Plot the solution
plt.figure()
plt.plot(sol.t, sol.y[0], '-', label='x1(t)')
plt.plot(sol.t, sol.y[1], ':', label='x2(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()
```

# Mod 6 - 4

```python
# Define the system of first-order differential equations
def system(t, y):
    dydt = np.zeros(2)
    u1 = 1 if t >= 1 else 0
    u2 = 1 if t >= 5 else 0
    dydt[0] = y[0] - 5*y[1] + np.exp(-0.02)*u1
    dydt[1] = 2*y[0] - 5*y[1] - 0.5*y[0] + 10*(1 - np.exp(-0.5*(t-5)))*u2
    return dydt

# Define the time points
t = np.linspace(0, 30, 1000)

# Define the initial conditions
y0 = [-7, 2]

# Solve the system of differential equations
sol = solve_ivp(system, [0, 30], y0, t_eval=t)

# Plot the solution
plt.figure()
plt.plot(sol.t, sol.y[0], '-', label='x1(t)')
plt.plot(sol.t, sol.y[1], '-', label='x2(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()

# Print the final values of x1(t) and x2(t)
print('x1(t) stabilizes at:', sol.y[0][-1])
print('x2(t) stabilizes at:', sol.y[1][-1])
```
