# 1

Express the solution of the given initial value problem as a sum of two oscillations. Primes denote derivatives with respect to time t. Graph the solution functionâ€‹ x(t) in such a way that you can identify and label its period.

x" + 16x = 48cos(2t)

x(0) = x'(0) = 0

```python
# 1 - 1

t = smp.symbols('t')
x = smp.Function('x')

# Define the differential equation
diff_eq = smp.Eq(x(t).diff(t, t) + 16*x(t), 48*smp.cos(2*t))

# Solve the differential equation
sol = smp.dsolve(diff_eq)

# Define the initial conditions
in_conds = {x(0): 0, x(t).diff(t).subs(t, 0): 0}

# Substitute the initial conditions into the solution
gen_sol = sol.subs(in_conds)

gen_sol
```

```python
# 1 - 1

# Substitute the initial conditions to solve for C1 and C2
C1, C2 = smp.symbols('C1 C2')

init_sol = sol.rhs.subs({x(t): 0, t: 0})
init_sol_diff = sol.rhs.diff(t).subs({x(t): 0, t: 0})

constants = smp.solve([init_sol, init_sol_diff], (C1, C2))

# Substitute the values of the constants back into the solution
specific_sol = sol.subs(constants)

specific_sol
```

```python
# 1 - 1

# Period of a function = distance (time interval) over which the function's pattern repeats
smp.periodicity(4*smp.cos(2*t) - 4*smp.cos(4*t), t)
```

```python
# 1 - 1

t_values = np.linspace(0, 2*np.pi, 1000)  # values of t for the graph
x_values = 4*np.cos(2*t_values) - 4*np.cos(4*t_values)  # corresponding values of x(t)

plt.figure(figsize=(10, 6))
plt.plot(t_values, x_values, label='x(t) = 4cos(2t) - 4cos(4t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Solution to the differential equation')
plt.legend(loc = 'lower right')
plt.grid(True)
plt.show()
```

# 2

Find the steady periodic solution x_sp (t) = Ccos(omega*t - alpha) of the given equation mx" + cx' + kx = F(t) with periodic forcing function F(t) of frequency omega. Then graph x_sp (t) together with (for comparison) the adjusted function F_1 (t) = (F(t)) / m(omega)

x" + 3x' + 33x = 10cos(5t)

```python
# 2 - 1

t = smp.symbols('t')
x = smp.Function('x')

diffeq = smp.Eq(x(t).diff(t, t) + 3*x(t).diff(t) + 33*x(t), 10*smp.cos(5*t))


sol = smp.dsolve(diffeq)
display(sol.rhs)

A = np.sqrt((150/289)**2 + (80/289)**2)
delta = np.arctan2(150/289, 80/289)
display(A)
display(delta)
print(f"{A}cos(5t - {delta})")
```

```python
# 2 - 1

t = smp.symbols('t')
x = smp.Function('x')

# Define the differential equation
diff_eq = smp.Eq(x(t).diff(t, t) + 3*x(t).diff(t) + 33*x(t), 10*smp.cos(5*t))

# Solve the differential equation
sol = smp.dsolve(diff_eq)

# Find the amplitude and phase shift
C = smp.sqrt(sol.rhs.args[0]**2 + sol.rhs.args[1]**2)
alpha = smp.atan2(sol.rhs.args[0], sol.rhs.args[1])

# Define the steady periodic solution
x_sp = C*smp.cos(5*t - alpha)

# Define the adjusted function
F_1 = 10*smp.cos(5*t) / (1*5)

# Create a range of t values
t_vals = np.linspace(0, 2*np.pi, 1000)

# Convert the symbolic functions to numeric functions
x_sp_numeric = smp.lambdify(t, x_sp, 'numpy')
F_1_numeric = smp.lambdify(t, F_1, 'numpy')

# Plot the steady periodic solution and the adjusted function
plt.figure(figsize=(10, 6))
plt.plot(t_vals, x_sp_numeric(t_vals), label='x_sp(t)')
plt.plot(t_vals, F_1_numeric(t_vals), label='F_1(t)')
plt.xlabel('t')
plt.ylabel('x(t), F_1(t)')
plt.title('Steady Periodic Solution and Adjusted Function')
plt.legend()
plt.grid(True)
plt.show()
```

A mass weighing 140 lbs (mass m = 4.375 in fps) is attached to the end of a spring that is stretched 6" by a force of 140 lbs. 

A force F_0 cos(omega*t) acts on the mass.

At what frequency (in hertz) will resonance oscillations occur? Neglect damping.

```python
# 3 - 1

# Define the symbols
m = 4.375  # mass in slugs
k = 140 / (6 / 12)  # spring constant in lb/ft

# Calculate the angular frequency
omega = smp.sqrt(k / m)

# Convert rad/s to Hz
omega_hz = omega / (2 * smp.pi)

omega_hz.evalf()
```

RL circuit. This linear first-order diff eq governs the current by:

Ll' + Rl = E(t)

Suppose L = 5, R = 20 Î©, and source E of emf is an AC generator supplying voltage of E(t) = 75cos(40t) V

Suppose the switch is initially in position 2, but is thrown to position 1 at time t = 0 so that l(0) = 0

Determine the subsequent inductor current, l(t)

```python
# 4 - 1

t = smp.symbols('t')
l = smp.Function('l')
C1 = smp.symbols('C1')

# Constants
L = 5
R = 20
E = 75*smp.cos(40*t)

# Diff eq
diffeq = smp.Eq(L*l(t).diff(t) + R*l(t), E)

# Solve
sol = smp.dsolve(diffeq)

# Apply initial conditions l(0) = 0
init_conds = {l(0): 0}

# General Solution
gen_sol = sol.subs(init_conds)
display(gen_sol)

# Specific Solution
C1_sol = smp.solve(gen_sol.rhs.subs(t, 0), C1)

gen_sol.subs(C1, C1_sol[0])
```

RC circuit. Resistor = (r ohms), capacitor (C farads), switch, source of emf, but no inductor. 

Charge Q = Q(t) on capacitor at time t satisfies the following linear first-order diff eq. 

R(dQ/dt) + (1/C)Q = E(t)

I(t) = Q'(t)

(a) = Find charge Q(t) and current I(t) in the RC circuit if E(t) = V_0 (constant voltage supplied by a battery) and the switch is closed at time t = 0, so that Q(0) = 0

(b) = Show that lim (t->smp.oo) Q(t) = (V_0)C and that lim(t->smp.oo) I(t) = 0

Id est: What are the solutions for Q(t) and I(t) that share a function of t in their definitions, f(t)?

```python
# Define the symbols
t, C, R, V_0, c_1 = smp.symbols('t C R V_0 c_1')

# Define the charge Q(t)
Q = c_1 * smp.exp(-t/(C*R)) + C*V_0

# Compute the current I(t)
I = smp.diff(Q, t)

# Substitute the initial condition Q(0) = 0 to find the constant c_1
c_1_val = smp.solve(Q.subs(t, 0), c_1)[0]

# Substitute the value of c_1 back into Q(t) and I(t)
Q = Q.subs(c_1, c_1_val) # Q(t)
I = I.subs(c_1, c_1_val) # I(t)

# A
display(Q)
display(I)

# B
# lim(t->smp.oo) = 0

# Since Q(t) is the product of V_0 (C) and 1 - f, lim(t->smp.oo) Q(t) = V_0 (C)
# Since I(t) is the product of ((V_0)/R) and f, lim(t->smp.oo) I(t) = 0
```
