Sketch a graph of the function f defined for all t by the given formula, and determine whether it is periodic. 

If so, find its smallest period:

f(t) = |sin2t|

```python
# Define the time variable
t = np.linspace(0, 4*np.pi, 1000)  # from 0 to 4pi with 1000 points in between

# Define the function
f = np.abs(np.sin(2*t))

# Plot the function
plt.figure(figsize=(10,6))
plt.plot(t, f)
plt.title('Plot of f(t) = |sin(2t)|')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(True)
plt.show()
```

```python
# Frequency of the function
B = 2

# Period of sin(2t)
T_sin = 2*np.pi / np.abs(B)

# Period of |sin(2t)|
T_abs_sin = T_sin / 2

print('The smallest period of the function is:')
T_abs_sin
```
